```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Yhills_July12_Analytics/main/titanic-training-data.csv")
```

```python
df.head()
```

```python
df.info
```

```python
df.dtypes
```

```python
sns.relplot(x="Pclass",y="Age", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",jitter=False, data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age", hue="Sex", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex", kind="swarm", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex", kind="box", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex", kind="violin", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex", kind="violin",split=True, data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex", kind="bar", data=df)
plt.show()
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex", kind="point", data=df)
plt.show()
```
